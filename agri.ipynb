```python
# Data Cleaning and Preprocessing
# Check for missing values
df.isnull().sum()
```

```python
#check for special(-1) values
(df == -1).sum()
```

```python
# replace -1 with NaN
df.replace(-1, np.nan, inplace=True)
```

```python
# replace NaN values with the mean of the respective column
df=df.fillna(df.mean(numeric_only=True))
```

```python
#check for missing values again
df.isnull().sum()
```

```python
# cleaning the column names
df.columns = (df.columns
.str.strip()
.str.lower()
.str.replace(' ', '_')
.str.replace(r'\([^)]*\)', '', regex=True)
.str.rstrip('_'))
# Display the cleaned column names
df.columns
```

```python
# removing column names with NaN
df = df.dropna(axis=1, how='all')
df = df.dropna(axis=0, how='all')
```

```python
# 10. Millet Production (Last 50y)
df['millet_production'] = (
    df['pearl_millet_production'] + 
    df['finger_millet_production'] +
    df['sorghum_production'] 
 ) # Combining pearl millet, finger millet, and sorghum production into a single column

# Group by year and sum each millet crop
millet_types = df.groupby('year')[[
    'sorghum_production',
    'pearl_millet_production',
    'finger_millet_production'
]].sum().sort_index().tail(50)

# Plot stacked area chart
ax = millet_types.plot.area(
    figsize=(12, 6),
    stacked=True,
    cmap='Set2'
)

# Overlay line outlines for each millet type
cumulative = millet_types.cumsum(axis=1)
for column in millet_types.columns:
    plt.plot(millet_types.index, cumulative[column], linewidth=0.8, color='black')

plt.title("Millet Production by Type (Sorghum, Pearl, Finger) – Last 50 Years")
plt.xlabel("Year")
plt.ylabel("Production (1000 tons)")
plt.tight_layout()
plt.show()
```

```python
# 11. Sorghum Production (Kharif and Rabi) by Region

# Group by state and sum kharif & rabi production
sorghum_by_region = df.groupby('state_name')[[
    'kharif_sorghum_production',
    'rabi_sorghum_production'
]].sum()

# Sort by total production
sorghum_by_region['total'] = sorghum_by_region.sum(axis=1)
sorghum_by_region = sorghum_by_region.sort_values('total', ascending=False).drop(columns='total')

# Select top 8 producing states
sorghum_by_region = sorghum_by_region.head(8).reset_index()

#Melt the data for Seaborn
sorghum_melted = sorghum_by_region.melt(id_vars='state_name', 
                                        value_vars=['kharif_sorghum_production', 'rabi_sorghum_production'],
                                        var_name='Season', value_name='Production')

# Replace season names for better display
sorghum_melted['Season'] = sorghum_melted['Season'].str.replace('sorghum_', '').str.replace('_production', '').str.title()

# Plot the bar chart using Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(data=sorghum_melted, x='state_name', y='Production', hue='Season', palette='Set2', edgecolor='black')

plt.title('Sorghum Production by Region (Kharif vs Rabi)', fontsize=14)
plt.xlabel('State')
plt.ylabel('Production (1000 tons)')
plt.xticks(rotation=45)
plt.legend(title='Season')
plt.tight_layout()
plt.show()

```

```python
# 13. Soyabean Production by Top 5 States and Yield Efficiency
soyabean_by_state = df.groupby('state_name')[[
    'soyabean_production', 'soyabean_area', 'soyabean_yield'
]].sum() # Grouping by state and summing soyabean production, area, and yield

top5 = soyabean_by_state.sort_values('soyabean_production', ascending=False).head(5).reset_index() # Resetting index to convert Series to DataFrame

# Plot: Bar for Production + Line for Yield
fig, ax1 = plt.subplots(figsize=(12, 6)) # Create a figure and axis for the bar plot

# Bar plot (production)
sns.barplot(
    data=top5, # Using the top 5 states for soyabean production
    x='state_name', # Use the correct column name for states
    y='soyabean_production', # Bar height represents soyabean production
    ax=ax1, # Specify the axis for the bar plot
    palette='Set2', # Using a color palette for better aesthetics
    edgecolor='black' # Adding black edges to the bars

)
ax1.set_ylabel('Soyabean Production (1000 tons)', color='black') # Set the y-axis label for the bar plot
ax1.set_xlabel('State') # Set the x-axis label
ax1.tick_params(axis='x', rotation=45) # Rotate x-axis labels for better readability
ax1.set_title('Top 5 Soyabean Producing States and Their Yield Efficiency') # Set the title of the plot
ax1.tick_params(axis='y', labelcolor='black') # Set the y-axis tick labels color for the bar plot

# Line plot (yield)
ax2 = ax1.twinx() # Create a second y-axis for the line plot
sns.lineplot(
    data=top5, 
    x='state_name',
    y='soyabean_yield',
    ax=ax2, # Specify the axis for the line plot
    color='darkblue',
    marker='o', # Adding markers to the line plot
    markersize=8, # Set the size of the markers
    linewidth=2, # Set the width of the line
    label='Yield Efficiency' 
)
ax2.set_ylabel('Yield (kg per hectare)', color='darkblue') # Set the y-axis label for the line plot
ax2.legend(loc='upper right') # Add a legend for the line plot

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 16. Rice vs Wheat Yield Across States

# Group by state and calculate average yield for rice and wheat
yield_by_state = df.groupby('state_name')[['rice_yield', 'wheat_yield']].mean()

# Select states based on average rice + wheat yield
yield_by_state['avg_total'] = yield_by_state.mean(axis=1)
yield_top = yield_by_state.sort_values('avg_total', ascending=False).drop(columns='avg_total').reset_index()


melted = yield_top.melt(id_vars='state_name', value_vars=['rice_yield', 'wheat_yield'],
                        var_name='Crop', value_name='Yield') # Melt the data for Seaborn grouped bar chart

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(data=melted, x='state_name', y='Yield', hue='Crop', palette='Set1', edgecolor='black')

plt.title('Rice vs Wheat Yield Across Top 10 States (kg/ha)', fontsize=14)
plt.xlabel('State')
plt.ylabel('Yield (kg per hectare)')
plt.xticks(rotation=45)
plt.legend(title='Crop')
plt.tight_layout()
plt.show()
```
